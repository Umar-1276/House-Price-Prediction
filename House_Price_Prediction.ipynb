Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

Load dataset

```python
housing_price_dataset = pd.read_csv('housing_price_dataset.csv', header=None)
```

```python
housing_price_dataset.head()
```

```python
housing_price_dataset.shape
```

```python
housing_price_dataset.describe()
```

Understanding the correlation between various features in the dataset

1. Positive Correlation

2. Negative Correlation

```python
# make first row the header (CSV was read with header=None)
housing_price_dataset.columns = housing_price_dataset.iloc[0]
housing_price_dataset = housing_price_dataset[1:].reset_index(drop=True)

# ensure numeric columns are numeric
num_cols = ['SquareFeet', 'Bedrooms', 'Bathrooms', 'YearBuilt', 'Price']
for c in num_cols:
    housing_price_dataset[c] = pd.to_numeric(housing_price_dataset[c], errors='coerce')

housing_price_dataset_numeric = housing_price_dataset.drop('Neighborhood', axis=1)
corelation = housing_price_dataset_numeric.corr()
display(corelation)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(corelation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and Target

```python
X = housing_price_dataset.drop(['Price', 'Neighborhood'], axis=1)
Y = housing_price_dataset['Price']
```

```python
print(X)
print(Y)
```

Splitting the data into Training data and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```

Prediction on Test Data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```
